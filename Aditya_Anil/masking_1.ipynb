```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

# Added Weights method white BG

```python
base_img=cv2.imread('images/2.jpg')
base_img=cv2.cvtColor(base_img,cv2.COLOR_BGR2RGB)
plt.imshow(base_img)
```

```python
base_img.shape
```

```python
white=cv2.imread('smoke_white.jpeg')
white=cv2.cvtColor(white,cv2.COLOR_BGR2RGB)
plt.imshow(white)
```

```python
white.shape
```

```python
overlay_white=white[400:599,200:388,:]
plt.imshow(overlay_white)
```

```python
plt.imshow(cv2.addWeighted(overlay_white,0.4,base_img,0.8,0))
```

# Added Weights Black BG


```python
black=cv2.imread('smoke_black.jpg')
black=cv2.cvtColor(black,cv2.COLOR_BGR2RGB)
plt.imshow(black)
```

```python
black.shape
```

```python
overlay_black=black[600:799,600:788,:]
plt.imshow(overlay_black)
```

```python
plt.imshow(cv2.addWeighted(overlay_black,0.5,base_img,1,0))
```

# Masking Method

```python
hsv_img=cv2.cvtColor(white,cv2.COLOR_RGB2HSV)
lower=np.array([0,0,0])
upper=np.array([254,254,254])
mask=cv2.inRange(hsv_img,lower,upper)
```

```python
plt.imshow(mask)
```

```python
plt.imshow(white)
```

```python
plt.imshow(cv2.bitwise_and(white,white,mask=mask))
```
